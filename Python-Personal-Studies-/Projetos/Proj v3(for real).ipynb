```python
from IPython.display import clear_output
#limpa tela
clear_output()

#Variaveis globais 
cont = 1.0
var = "X"
l = [[1,2,3],[4,5,6],[7,8,9]]
list1,list2,list3=[x for x in l[0]],[x for x in l[1]],[x for x in l[2]]
terminou = False

#Função paa mostrar o board
def display():
    print("{a}\n{b}\n{c}\n".format(a=list1,b=list2,c=list3))
    return  

#Função para capturar a escolha do jogador
def escolha():
    posição = int(input("Escolha um dos numeros do tabuleiro abaixo: "))
    #print ("Você escolheu %s" %(posição))
    return posição

#Usar X ou O ?
def marcador():
    global cont
    global var
    if cont % 2 == 0:
        var = "O"
    else:
        var ="X"
    cont +=1
    return var

#Teste de condições para vitoria
def vitoria_S2():
    global cont
    global list1,list2,list3
    global terminou

    if list1[0] == list1[1] == list1[2]:
        print ("Fim de jogo, vitoria de: %s" %(list1[0]))
        display()
        terminou = True
        
    elif list2[0] == list2[1] == list2[2]:
        print ("Fim de jogo, vitoria de: %s" %(list2[0]))
        display()
        terminou = True
        
    elif list3[0] == list3[1] == list3[2]:
        print ("Fim de jogo, vitoria de: %s" %(list3[0]))
        display()
        terminou = True
        
    elif list1[0] == list2[0] == list3[0]:
        print ("Fim de jogo, vitoria de: %s" %(list1[0]))
        display()
        terminou = True
        
    elif list1[1] == list2[1] == list3[1]:
        print ("Fim de jogo, vitoria de: %s" %(list1[1]))
        terminou = True
        
    elif list1[2] == list2[2] == list3[2]:
        print ("Fim de jogo, vitoria de: %s" %(list1[2]))
        display()
        terminou = True
        
    elif list1[0] == list2[1] == list3[2]:
        print ("Fim de jogo, vitoria de: %s" %(list1[0])) 
        display()
        terminou = True
        
    elif list1[2] == list2[1] == list3[0]:
        print ("Fim de jogo, vitoria de: %s" %(list1[2]))
        display()
        terminou = True
        
    elif cont >9:
        display()
        terminou = True
        
    return

#Função para colocar X ou O no board
def inserir(item,var):
    global list1,list2,list3
    if item <=3:
        if item in list1:
            list1[list1.index(item)] = var
    elif 6 >= item >=4:
        if item in list2:
            list2[list2.index(item)] = var
    elif 9 >= item >=7:
        if item in list3:
            list3[list3.index(item)] = var
    else:
        print("Você escolheu um numero invalido, seu burro!")
    return        

def main():
    print("Bem vindo ao jogo da velha")
    while not terminou :
        display()
        inserir(escolha(),marcador())
        vitoria_S2()

    print ("Obrigado")
    return 
```

```python
main()
```
